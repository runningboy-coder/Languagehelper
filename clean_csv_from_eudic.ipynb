## 提取词性

```python
vocabulary.drop(["#","笔记"],axis=1,inplace=True)
```

```python
vocabulary.head(10)
```

```python
vocabulary["解释"][0]
```

```python
extract_re = r"([a-z]*)\. "
vocabulary["词性"] = vocabulary["解释"].str.extract(extract_re)
vocabulary["词性"].head(10)
```

```python
vocabulary["词性"].value_counts()
```

```python
word_properties = ['v','vt','vi','n','prep','adv','adj','conj','abbr']
def clean_abnormal_value(content):
    if content not in word_properties:
        return None
    else:
        return content

vocabulary["词性"].apply(clean_abnormal_value)
vocabulary.head(10)
```

```python
vocabulary["clean_meaning"] = vocabulary["解释"].str.extractall(r"(\w)")
vocabulary.sample(10)
```
